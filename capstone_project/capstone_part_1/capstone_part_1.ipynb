```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.magicbricks.com')
```

```python
search=driver.find_element(By.XPATH,"/html/body/section/section/div/div[1]/div[3]/div[4]")
search.click()
```

```python
name=[]
name_of_building=[]
Super_Area=[]
Under_Construction=[]
Transaction=[]
Furnishing=[]
Bathroom=[]
price=[]
builder=[]
```

```python
name_tags = driver.find_elements(By.XPATH, '//h2[@class="mb-srp__card--title"]')
# Check if any elements were found
if name_tags:
    for i in name_tags:
        titles = i.text
        name.append(titles)
else:
    name.append(None)  # Append None to the list if no elements were found
```

```python
try:
    # Assuming you have a way to scrape these values
    buildings = driver.find_elements(By.XPATH, '//span[@class="mb-srp__card__developer--name--highlight"]')
    areas = driver.find_elements(By.XPATH, '//div[@class="mb-srp__card__summary open"]/div/div[1]')
    construction_status = driver.find_elements(By.XPATH, '//div[@class="mb-srp__card__summary open"]/div/div[2]')
    transactions = driver.find_elements(By.XPATH, '//div[@class="mb-srp__card__summary open"]/div/div[3]')
    furnishings = driver.find_elements(By.XPATH, '//div[@class="mb-srp__card__summary open"]/div/div[4]')
    bathrooms = driver.find_elements(By.XPATH, '//div[@class="mb-srp__card__summary open"]/div/div[5]')
    prices = driver.find_elements(By.XPATH, '//div[@class="mb-srp__card__price--amount"]')
    builders = driver.find_elements(By.XPATH, '//div[@class="mb-srp__card__ads--name"]')

    # Populate the lists
    for building in buildings:
        if building.text.strip():
            name_of_building.append(building.text)
        else:
            name_of_building.append(None)
        
    for area in areas:
        if area.text.strip():
            Super_Area.append(area.text)
        else:
            Super_Area.append(None)
        
    for status in construction_status:
        if status.text.strip():
            Under_Construction.append(status.text)
        else:
            Under_Construction.append(None)
        
    for transaction in transactions:
        if transaction.text.strip():
            Transaction.append(transaction.text)
        else:
            Transaction.append(None)
        
    for furnishing in furnishings:
        if furnishing.text.strip():
            Furnishing.append(furnishing.text)
        else:
            Furnishing.append(None)
        
    for bathroom in bathrooms:
        if bathroom.text.strip():
            Bathroom.append(bathroom.text)
        else:
            Bathroom.append(None)
        
    for price_item in prices:
        if price_item.text.strip():
            price.append(price_item.text)
        else:
            price.append(None)
        
    for builder_item in builders:
        if builder_item.text.strip():
            builder.append(builder_item.text)
        else:
            builder.append(None)

except NoSuchElementException:
    # Handle the case where elements are not found
    name_of_building.append(None)
    Super_Area.append(None)
    Under_Construction.append(None)
    Transaction.append(None)
    Furnishing.append(None)
    Bathroom.append(None)
    price.append(None)
    builder.append(None)
```

```python
print(len(name),len(name_of_building),len(Super_Area),len(Under_Construction),len(Transaction),len(Furnishing),len(Bathroom),len(price),len(builder))
```

```python
import pandas as pd



# Create a list of all lists
lists = [name, name_of_building, Super_Area, Under_Construction, 
         Transaction, Furnishing, Bathroom, price, builder]

# Find the maximum length
max_length = max(len(lst) for lst in lists)

# Pad lists with None to make them the same length
padded_data = {
    'name': name + [None] * (max_length - len(name)),
    'name_of_building': name_of_building + [None] * (max_length - len(name_of_building)),
    'Super_Area': Super_Area + [None] * (max_length - len(Super_Area)),
    'Under_Construction': Under_Construction + [None] * (max_length - len(Under_Construction)),
    'Transaction': Transaction + [None] * (max_length - len(Transaction)),
    'Furnishing': Furnishing + [None] * (max_length - len(Furnishing)),
    'Bathroom': Bathroom + [None] * (max_length - len(Bathroom)),
    'price': price + [None] * (max_length - len(price)),
    'builder': builder + [None] * (max_length - len(builder)),
}

# Create DataFrame
df = pd.DataFrame(padded_data)

# Display the DataFrame
df.head(30)
```

```python
df.to_csv('data1.csv', index=False)
```
